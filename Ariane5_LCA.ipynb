```python
#objective of the mission is to launch 5000 kg of payload into orbit
#so we need to create a functional unit that represents the launch of 1 kg of payload
func_unit_launcher = {   'name': 'launch of ariane 5',
    'unit': 'kilogram',
    'location': 'RER',
    'exchanges': [{
        'input': ("components", "launcher"),
        'amount': 1. / 5000,  # 1 launch delivers 5 tons de payload, so we divide by 5000 to get per kg
        'type': 'technosphere'
    }]
}

na = components.new_activity('launch of ariane 5 per kg payload', **func_unit_launcher)

try:
    na.save()
except IntegrityError:
    pass

if not na.exchanges():
    for exc in func_unit_launcher["exchanges"]:
        na.new_exchange(**exc).save()

activity = components.get('launch of ariane 5 per kg payload')
```

```python
#performing the LCA using the ReCiPe Midpoint (H) V1.13 method for GWP100 per kg of payload
method_key = ('ReCiPe Midpoint (H) V1.13', 'climate change', 'GWP100')
#method = bw.methods[method_key]
lca = bw.LCA({activity: 1}, method_key)
lca.lci()
lca.lcia()
outputs = {} 
outputs['GWP'] = lca.score 
print(f"GWP100 per kg payload: {outputs['GWP']:.2f} kg CO₂-eq")

#aligns with ESA LCA for Ariane 5 : 900 to 1200 kg CO2 eq/kg of payload
```

```python
#scale to 5 tons payload
per_launch = outputs['GWP'] * 5000  # tons of CO₂-eq per launch
print(f"GWP per launch (5t): {per_launch:.2f} kg CO₂-eq -> {per_launch / 1000:.2f} tons CO₂-eq per launch with 5t payload")
```

```python
# loop through all ReCiPe Midpoint (H) V1.13 methods
method_keys = []
for key in bw.methods:
    if key[0] == 'ReCiPe Midpoint (H) V1.13':
        method_keys.append(key)

# store and print impacts per kg and per launch
print("impact categories  | per kg payload | per 5t launch):")
for method in method_keys:
    lca.switch_method(method)
    lca.lcia()
    impact_name = f"{method[1]} - {method[2]}"
    per_kg = lca.score
    per_launch = per_kg * 5000
    print(f"{impact_name:40s} | {per_kg:10.2f} per kg | {per_launch/1000:10.2f} per launch (t)")
```

```python
#multi impact LCIA

lca = bw.LCA({activity: 1}, method_keys[0])  # LCA object using the first method (e.g., GWP100)
lca.lci()                                    # LCI calculation
lca.decompose_technosphere()                 # optimise for calculations
lca.lcia()                                   

characterization_matrices = []
for method in method_keys:
    lca.switch_method(method)
    lca.redo_lcia()
    print("Score for impact '{0}' is {1:.3e} {2}".format(method[1], lca.score, bw.methods[method]['unit']))
```

```python
# top 10 contributors to GWP100

from brightway2 import get_activity

method_key = ('ReCiPe Midpoint (H) V1.13', 'climate change', 'GWP100')
lca = bw.LCA({activity: 1}, method_key)
lca.lci()
lca.lcia()

# Compute and sort contributions
contributions = [
    (key, (lca.characterization_matrix.multiply(lca.inventory[:, i])).sum())
    for key, i in lca.activity_dict.items()
]
top10 = sorted(contributions, key=lambda x: x[1], reverse=True)[:10]

print("Top 10 contributors to GWP100:")
for key, score in top10:
    name = get_activity(key)['name']
    rel = 100 * score / lca.score
    print(f"{name[:40]:40s} | {score:10.2f} | {rel:5.1f}% of total")
```
